### 1- Input and Output Exercise

### Exercise 1: Accept numbers from a user

```python
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))
print(f"The total {num1 * num2}")
```

### Exercise 2: Display three strings: "Name", "Is", "James" as "Name*Is*James"

```python
name = "Name"
IS = "Is"
James = "James"

print(f"{name}**{IS}**{James}")
```

### Exercise 3: Convert a decimal number to octal using print() output formatting

```python
num = 8
print(f"The octal number of decimal number {num} is {oct(num)[2:]}")
```

### Exercise 4: Display a float number with 2 decimal places using print()

```python
num = 458.541315
print(round(num, 2))
```
