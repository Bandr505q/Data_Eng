### Exercise 2: Arrange string characters such that lowercase letters come first

```python
s = "PyNaTive"

lower = ""
upper = ""

for ch in s:
    if ch.islower():
        lower = lower + ch
    else:
        upper = upper + ch

result = lower + upper
print(result)
```

### Exercise 3: Count all letters, digits, and special symbols from a given string

```python
s = "P@#yn26at&i5ve"

chars = 0
digits = 0
symbols = 0

for ch in s:
    if ch.isalpha():
        chars = chars + 1
    elif ch.isdigit():
        digits += 1
    else:
        symbols += 1

print("Total counts of chars, digits, and symbols")
print("Chars =", chars)
print("Digits =", digits)
print("Symbols =", symbols)
```

### Exercise 4: String character balance test

```python
s1 = "Yn"
s2 = "PYnative"

balanced = True

for ch in s1:
    if ch not in s2:
        balanced = False
        break

print(balanced)
```

### Exercise 6: Calculate the sum and average of the digits present in a string

```python
s = "PYnative29@#8496"

total = 0
count = 0

for ch in s:
    if ch.isdigit():
        total += int(ch)
        count += 1

average = total / count

print("Sum is:", total, "Average is:", average)
```

### Exercise 7: Write a program to count occurrences of all characters within a string

```python
s = "Apple"

counts = {}

for ch in s:
    if ch in counts:
        counts[ch] += 1
    else:
        counts[ch] = 1

print(counts)
```

### Exercise 10: Remove empty strings from a list of strings

```python
str_list = ["Emma", "Jon", "", "Kelly", None, "Eric", ""]

print("Original list of string ")
print(f"{str_list}\n")

result = []

for item in str_list:
    if item:
        result.append(item)

print("After removing empty strings")
print(result)
```

### Exercise 11: Replace each special symbol with # in the following string

```python
s = "/Jon is @developer & musician!!"

result = ""
for ch in s:
    result += ch if ch.isalnum() or ch == " " else "#"

print(result)
```
